    <h1>Descriptive Statistics</h1>
<article id="#normal-dist">
    <h3>Normal Distribution</h3>
    <p>Data needs to be distributed normally in order to perform a majority of statistical tests, this normailty is usually shown as a "bell" or "Gaussian" curve.</p>
    <p>Data is "normally distributed" if each side is a reflection of itself, bell shaped, centered and unimodal (single peak).</p>
    <img src="https://miro.medium.com/max/700/0*N7NAMAeUN41oToZ-.png" alt="Example of bell curve graph">
</article>
<article id="#central-tend">
    <h3>Central Tendancy</h3>
    <p>Central Tendency: mean, median and mode. Mean: average. Mode: value that occurs most often. Median: value the occurs in the midle of dataset (50th percentile).</p>
</article>
<article id="#measure-var">
    <h3>Measures of Variability</h3>
    <p>Range, IQR, StdDev. Range: max_val - min_val. IQR: statistical dispersion between upper and lower quartiles. StdDev: how spread apart is the data.</p>
    <p>Tukey Outlier</p>
    <ol>
        <li>Find IQR (75% - 25% or q3 - q1)</li>
        <li>bounds = IQR * 1.5</li>
        <li>Range to look for outliers: (X &lt; q1 - bounds) || (X &gt; q3 + bounds)</li>
    </ol>
    <div style="margin:0 37% 0 37%; border:2px solid #FEFBEA; padding: 2px;">
        <p>Sample Std. Dev. = $\sqrt{\frac{1}{n-1}\sum_{i=0}^n(x-\bar{x})^2}$, where $\bar{x} =$ mean of sample </p>
        <p>Population Std. Dev. = $\sqrt{\frac{1}{n}\sum_{i=0}^n(x-\mu)^2}$, where $\mu =$ mean of population </p>
        <p>Mean Deviation / Mean Absolute Deviation = $\frac{1}{n}\sum_{i=0}^n|X_{i} - \bar{x}|$</p>
        <p>Variance: (standard deviation)$^2$</p>
    </div>
    <p>Z-score: how many std dev's above or below the mean a specific point lies</p>
    <img src="https://miro.medium.com/max/700/0*1A7jzRbqRHafgfiP.jpg" alt="data distribution percentages by region">
</article>
<article id="#mod-skew">
    <h3>Modality and Skewness</h3>
    <img src="https://miro.medium.com/max/700/0*m_Fd3Opt6L70LiYS.png" alt="Examples of modality"><br>
    <figure>
        <img src="https://miro.medium.com/max/600/0*F1mkGYUbmqtZzLKF.jpg" alt="Example of skewness in distributions">
        <figcaption>The Pearson skewness coefficient provides quick estimation of distributions symmetry. To compute in pandas, use .skew() function</figcaption>
    </figure>
    <p>Pearson First Coefficient of Skewness (Mode Skewness): $\frac{Mean - Mode}{Standard Deviation}$</p>
    <p>Pearson Second Coefficient of Skewness (Median Skewness): $\frac{3(Mean - Median)}{Standard Deviation}$
</article>
<article id="#kurtosis">
    <h3>Kurtosis</h3>
    <p>Kurtosis: measures where your dataset is heavy-tailed or light-tailed compared to a normal distribution. Datasets with high kurtosis have heavy tails and more outliers, low kurtosis have light tails and fewer outliers.</p>
    <p>A histogram is a good way to show both the skewness and kurtosis of a dataset as it allows one to quickly identify if something is wrong with your data.</p>
    <p>Mathematical way to meaure for kurtosis is to use the fisher measurement of kurtosis.</p>
    <p><b>3 types of kurtosis:</b></p>
    <ul>
        <li>Mesokurtic: kurtosis around zero (0)</li>
        <li>Platykurtic: negative kurtosis and tails very thin</li>
        <li>Leptokurtic: kurtosis greater than 3 and fat tails; more extreme values, smaller std dev</li>
    </ul>
    <p>Pandas has a built in kurtosis() function.</p>
    <figure>
        <img src="https://miro.medium.com/max/375/1*_N2t_tWB7y3_N0ilzgTQng.jpeg" alt="3 types of kurtosis">

<div id=#anova style="background-color: #08415C; padding-left: 5px">
    <h1>One Way ANOVA</h1>
    <article id="#hypothesis">
        <h3>Define Hypothesis</h3>
        <p>H0: All levels or groups in set have equal variance<br>H1: At least one group is different</p>
    </article>
    <article id="#sum-squares">
        <h3>Calculate Sum of Squares</h3>
        <p>Sum of squares is the statistical technique used to determine the dispersion in data points. It is a measure of deviation.</p>
        <div style="margin:0 37% 0 37%; border:2px solid #FEFBEA; padding: 2px">
            <p>Sum of Squares = $\sum_{i=0}(X_{i} - \bar{X})^2$<br><b>where:</b><br>$X_{i} =$ The $i^{th}$ item in the set<br>$\bar{X} =$ The mean of all items in the set<br>$(X_{i} - \bar{X}) =$ The deviation of each item from the mean</p>
        </div>
        <p>Total Sum of Squares<br>Distance between each observed point x from the grand mean xbar  is x-xbar; if you calculate this distance between each data pount, square each distance and add up all the squared distances you get:</p>
        <div style="margin-left:37%;">
            <p>SST = $\sum(x_{i} - \bar{x})^2$</p>
        </div>
        <p>Between the Sum of Squares<br>The distance between each group average value g from grand means xbar is g-xbar. Doing similar to total sum of squares we get:</p>
        <div style="margin-left:37%;">
            <p>SSB = $\sum(g_{i} - \bar{x})^2$</p>
        </div>
        <p>Within the Sum of Squares<br>The distance between each observed value within the group x from the group-mean g is given as x-g; similar to sum of squares we get:</p>
        <div style="margin-left:37%;">
            <p>SSE = $\sum(x_{i} - g)^2$</p>
        </div>
    </article>
    <article id="#dof">
        <p>We already discussed what is the definition of degrees of freedom now we will calculate for between groups and within groups.</p>
        <ol>
            <li>Since we have 3 groups ( mother, father, other) degrees of freedom for Between groups can be given as (3–1) = 2</li>
            <li>Having 18 samples in each group, Degrees of freedom for within groups will be the sum of degrees of freedom of all groups that is (18–1) + (18–1) + (18–1) = 51</li>
        </ol>
    </article>
    <article id="#f-value">
        <p>F-Value<br>Since we are comparing the variance between the groups and variance within the groups. The F value is given as:</p>
        <div style="margin-left:37%;">
            <p>F = (SSB/df_b) / (SSW/df_w)</p>
        </div>
    </article>
</div>

<article id="#anova-example" style="font-size: 20px;">ANOVA Example</article>

```python
import pandas as pd
```

```python
# Data creation
A = [12.6, 12, 11.8, 11.9, 13, 12.5, 14]
B = [10, 10.2, 10, 12, 14, 13]
C = [10.1, 13, 13.4, 12.9, 8.9, 10.7, 13.6, 12]
all_scores = A + B + C
company_names = (['A'] * len(A)) +  (['B'] * len(B)) +  (['C'] * len(C))
data = pd.DataFrame({'company': company_names, 'score': all_scores})
```

```python
data.head()
```

```python
# Descriptives
data.groupby('company').mean()
```

```python
# A 1-Way ANOVA Using StatsModels
import statsmodels.api as sm
from statsmodels.formula.api import ols
lm = ols('score ~ company',data=data).fit()
table = sm.stats.anova_lm(lm)
print(table)
```

```python
# 1-Way ANOVA from scratch

# compute overall mean
overall_mean = data['score'].mean()
print(f'overall mean: {overall_mean}')
```

```python
# compute Sum of Squares Total
data['overall_mean'] = overall_mean
ss_total = sum((data['score'] - data['overall_mean'])**2)
print(f'sum of squares total: {ss_total}')
```

```python
# compute group means
group_means = data.groupby('company').mean()
group_means = group_means.rename(columns = {'score': 'group_mean'})
group_means
```

```python
# add group means and overall mean to the original data frame
data = data.merge(group_means, left_on = 'company', right_index = True)
```

```python
# compute Sum of Squares Residual
ss_residual = sum((data['score'] - data['group_mean'])**2)
ss_residual
```

```python
# compute Sum of Squares Model
ss_explained = sum((data['overall_mean_x'] - data['group_mean'])**2)
ss_explained
```

```python
# compute Mean Square Residual
n_groups = len(set(data['company']))
n_obs = data.shape[0]
df_residual = n_obs - n_groups
ms_residual = ss_residual / df_residual
ms_residual
```

```python
# compute Mean Square Explained
df_explained = n_groups - 1
ms_explained = ss_explained / df_explained
ms_explained
```

```python
# compute F-Value
f = ms_explained / ms_residual
f
```

```python
# compute p-value
import scipy.stats
p_value = 1 - scipy.stats.f.cdf(f, df_explained, df_residual)
p_value
```

<p>Since our p-value of 0.45 is larger that 0.05, we cannot reject our null hypothesis and cannot accept our alternative. Even though the three sample means are different, we do not have a statistically significant difference.</p>
<p>This means that with the observed data, there is not enough evidence to assume a general difference in delivery times of the three companies.</p>

    <h1>Performance Metrics</h1>
<article id="#accuracy">
    <h3><b>Accuracy</b></h3>
    <p>The ratio of correctly labeled subjects to the whole pool of subjects.</p>
    <p>Accuracy answers the following question: How many students did we correctly label out of all the students?</p>
    <div style="margin-left:37%;">
        <p>Accuracy = $\frac{TP + TN}{TP + FP + FN + TN}$</p>
    </div>
    <p>Numerator: all correctly labeled subjects (all trues)<br>Denominator: all subjects</p>
</article>
<article id="#precision">
    <h3><b>Precision</b></h3>
    <p>The ration of the correctly positive labeled by our program to all positive laveled.</p>
    <p>Precision answers the following question: How many of those who we labaled as diabetic are actually diabetic?</p>
    <div style="margin-left:37%;">
        <p>Precision = $\frac{TP}{TP + FP}$</p>
    </div>
    <p>Numerator: postive labeled subjects<br>Denominator: all positive labeled by our program</p>
</article>
<article id="#recall">
    <h3><b>Recall (aka Sensitivity)</b></h3>
    <p>The ratio of correctly positive labeled by our program to all true positives.</p>
    <p>Recall answers the following question: Of all the people who are diabetic, how many of those we correctly predict?</p>
    <div style="margin-left:37%;">
        <p>Recall = $\frac{TP}{TP + FN}$</p>
    </div>
    <p>Numerator: positively labeled subject<br>Denominator: all people who are actually positive</p>
</article>
<article id="#f1">
    <h3><b>F1 Score (aka F-Score / F-Measure)</b></h3>
    <p>F1 Score considers both precision and recall. It is called the harmonic mean of the precision and recall.</p>
    <p>F1 score is best if there is some sort of balance between precision and recall in the system.</p>
    <div style="margin-left:37%;">
        <p>F1-Score = $2*(\frac{Recall * Precision}{Recall + Precision})$</p>
    </div>
</article>
<article id="#specificity">
    <h3><b>Specificity</b></h3>
    <p>The correctly negatively labeled items by the program to all who are actually healthy.</p>
    <p>Specificity answers the following question: Of all the people who are healthy, how many of those did we correctly predict?</p>
    <div style="margin-left:37%;">
        <p>Specificity = $\frac{TN}{TN + FP}$</p>
    </div>
    <p>Numerator: negatively labeled healthy people<br>Denominator: all people who are healthy in reality</p>
</article>
<article id="#which-metric">
    <h3><b>Which Metric to Use?</b></h3>
    <p>Accuracy is a great measure but only when you have symmetric datasets, <b>false negatives and false positives have similar costs</b>.</p>
    <p>F1 is used when the cost of false positives and false negatives are different (best for uneven class distribution).</p>
    <p>Precision is how sure you are of your true positives while recall is how sure you are that you are not missing any positives.</p>
    <p>Choose recall if the idea of false positives is better than false negatives (like with diagnosing for disease).</p>
    <p>Choose precision if you want to be more confident of your true positives (ex. spam emails, you'd rather some make it through the filter than miss important emails)</p>
    <p>Choose specificity if you want to cover all true negatives; you do not want any false alarms - false positives (ex. drug testing; you don't want to send people to jail by accident)</p>
    <p><b>Bottom line is:</b></p>
    <p>— Accuracy value of 90% means that 1 of every 10 labels is incorrect, and 9 is correct.<br>
    — Precision value of 80% means that on average, 2 of every 10 diabetic labeled student by our program is healthy, and 8 is diabetic.<br>
    — Recall value is 70% means that 3 of every 10 diabetic people in reality are missed by our program and 7 labeled as diabetic.<br>
    — Specificity value is 60% means that 4 of every 10 healthy people in reality are miss-labeled as diabetic and 6 are correctly labeled as healthy.</p>
</article>

    <h1>Types of Distance</h1>
    <article id="#euclidean">
        <h3>Euclidean Distance</h3>
        <p>Euclidean distance is what most people commonly think of when it comes to measuring distance; it is the 'straight line' distance between two points using x and y coordinates.</p>
        <p>$dist(A,B) = \sqrt{(x_{A} - x_{B})^2 + (y_{A} - y_{B})^2}$</p>
        <img src="https://miro.medium.com/max/1524/1*J2bK-UKhrW1Ill5EyAxXOQ.png" alt="Euclidean distance example">
    </article>
    <article id="#manhattan">
        <h3>Manhattan Distance (L1)</h3>
        <p>The L1 Distance, also called the Cityblock Distance, the Manhattan Distance, the Taxicab Distance, the Rectilinear Distance or the Snake Distance, does not go in straight lines but in blocks.</p>
        <p>$dist(A,B) = |x_{A} - x_{B}| + |y_{A} - y_{B}|$</p>
        <img src="https://miro.medium.com/max/700/1*88uZae0Utf7kavhQFvMqaw.png" alt="Cityblock distance example">
    </article>
    <article id="#chessboard">
        <h3>Chessboard Distance (Chebyshev)</h3>
        <p>The Chebyshev Distance is also called the Chessboard Distance or the L Infinity Distance. The most intuitive understanding of the Chebyshev distance is the movement of the King on a chessboard: it can go one step in any direction (up, down, left, right and verticals).</p>
        <p>$dist(A,B) = max(|x_{A} - x_{B}|, |y_{A} - y_{B}|)$</p>
